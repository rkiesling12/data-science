After collecting all of the data, I will summarize it. I will take the highest rank for each book, the maximum number of weeks on the list, and the first value for most of the other categories.

```python
import pandas as pd
fic_df = pd.read_csv('nyt_fic.csv')
```

I am curious whether any of the books receieved a dagger, which indicates that its spot on the bestseller list was primarily due to bulk sales.

```python
fic_df['dagger'].unique()
```

None of the bestsellers within this time period were marked with a dagger.

```python
books_grouped = fic_df.groupby('title')
combined_ranks = fic_df.groupby('title').agg( { 'rank': min, 'weeks_on_list': max, 'asterisk': max, 'dagger': max,
                                               'primary_isbn10': 'first', 'primary_isbn13': 'first', 'publisher': 'first',
                                               'description': 'first', 'price': max, 'author': 'first'}
)
```

After combining the books, I will look at how many bestsellers there were during this time period. There were 2442 books that were on the fiction bestseller list for at least one week during this period.

```python
len(combined_ranks)
```

```python
combined_ranks.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(combined_ranks['weeks_on_list'])
plt.show()
```

We can see from the plot above that most of the books only stay on the best sellers list for a short period of time, but a few have stayed on the list for over a year.
